```python
import pandas as pd
from sklearn.model_selection import train_test_split
from sklearn.preprocessing import OneHotEncoder
from sklearn.impute import SimpleImputer
from sklearn.preprocessing import StandardScaler
from sklearn.neighbors import KNeighborsClassifier

# import warnings filter
from warnings import simplefilter
# ignore all future warnings
simplefilter(action='ignore', category=FutureWarning)

#Cargar el conjunto de datos
datos = pd.read_csv('DataSetCovid/Raw-Data.csv')
# detectar valores faltantes
print(datos.isna().sum())

datos.head()


```

```python
# imputar los valores faltantes con la moda
imputer = SimpleImputer(strategy='most_frequent')
datos[['Country', 'Age', 'Gender', 'Symptoms', 'Experiencing_Symptoms', 'Severity','Contact']] = imputer.fit_transform(datos[['Country', 'Age', 'Gender', 'Symptoms', 'Experiencing_Symptoms', 'Severity','Contact']])

#datos.head(17)

# separar los datos en variables independientes y dependiente
X = datos.iloc[:, :-1].values
y = datos.iloc[:, -1].values

# codificar variables categ√≥ricas con One-Hot
encoder = OneHotEncoder(handle_unknown='ignore')
X = encoder.fit_transform(X).toarray()

# normalizar los datos
sc = StandardScaler()
X = sc.fit_transform(X)

# dividir los datos en conjuntos de entrenamiento y prueba
X_train, X_test, y_train, y_test = train_test_split(X, y, test_size=0.2, random_state=0)

# aplicar el algoritmo KNN con k=3
classifier = KNeighborsClassifier(n_neighbors=3)
classifier.fit(X_train, y_train)

# predecir los resultados del conjunto de prueba
y_pred = classifier.predict(X_test)

# evaluar el modelo
from sklearn.metrics import confusion_matrix, accuracy_score
cm = confusion_matrix(y_test, y_pred)
print(cm)
print('Exactitud: ', accuracy_score(y_test, y_pred))
```
